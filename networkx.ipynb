```python

def generate_dots_coordi(i,cent_coordix,cent_coordiy):
  num_dot = []
  x = cent_coordix
  y = cent_coordiy
  coordi = []
  for i in range(i,i+5):
    num_dot.append(i)
    if i%5 == 0:
      coordi_temp = (x,y)
      coordi.append(coordi_temp)
    if i%5 == 1:
      coordi_temp = (x-1,y-1)
      coordi.append(coordi_temp)
    if i%5 == 2:
      coordi_temp = (x+1,y-1)
      coordi.append(coordi_temp)
    if i%5 == 3:
      coordi_temp = (x+1,y+1)
      coordi.append(coordi_temp)
    if i%5 == 4:
      coordi_temp = (x-1,y+1)
      coordi.append(coordi_temp)
  pos_format = zip(num_dot,coordi)
  pos_dict = dict(pos_format)
  return pos_dict,num_dot

print(generate_dots_coordi(5,1,3.5))
```

```python
def generate_edges(i):
  center_dot = i
  cent_edgelist = []
  margin_edgelist = []
  edge_list = []
  for i in range(i,i+4):
    cent_edge = (center_dot,i+1)
    cent_edgelist.append(cent_edge)
    if i-center_dot < 3:
      margin_edge = (i+1,i+2)
      margin_edgelist.append(margin_edge)
    else:
      margin_edge = (i+1,i-2)
      margin_edgelist.append(margin_edge)
    edge_list = cent_edgelist + margin_edgelist
  return edge_list


print(generate_edges(0))
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def plot_n0_graph(cent0_point,cent_coordix,cent_coordiy):
  G = nx.Graph()
  generate_re = generate_dots_coordi(cent0_point,cent_coordix,cent_coordiy)
  pos = generate_re[0]
  node_numdot = generate_re[1]
  nx.draw_networkx_nodes(G, pos, node_size=200, nodelist=node_numdot,node_color="black")

  nx.draw_networkx_edges(G, pos, edgelist=generate_edges(0),edge_color="grey")

  plt.tight_layout()
  plt.axis("off")
  plt.show()

plot_n0_graph(0,1,1)
```
